```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score,f1_score, roc_auc_score, auc, log_loss, roc_curve
```

```python
results_df=pd.read_csv("test_result.csv")
results_df.head()
```

```python
y_actual=results_df['label']
y_pred=results_df['prediction']
```

```python
y_predict_proba=[]
ant_proba=[]
bee_proba=[]
cat_proba=[]
dog_proba=[]
for idx,row in results_df.iterrows():
    ant_proba.append(row[3])
    bee_proba.append(row[4])
    cat_proba.append(row[5])
    dog_proba.append(row[6])
    y_predict_proba.append([row[3],row[4],row[5],row[6]])
```

## Confusion Matrix

```python
def view_confusionmatrix(predictions, actuals):  
    cf_matrix = confusion_matrix(predictions, actuals)
    df_cm = pd.DataFrame(cf_matrix, columns=np.unique(actuals), index = np.unique(actuals))    
    df_cm.index.name = 'Actual'
    df_cm.columns.name = 'Predicted'
    sns.set(font_scale=1.4)
    sns.heatmap(df_cm, cmap="Blues", annot=True,fmt='g',annot_kws={"size": 16})
    return
view_confusionmatrix(y_pred, y_actual)
```

## F1 Score

```python
f1score=f1_score(y_true=y_actual,y_pred=y_pred,average='micro')
f1score=round(f1score,2)
print(f"F1 score: {f1score}")
```

## LogLoss

```python
logloss=log_loss(y_actual, y_predict_proba)
print(f"Multi class log loss: {round(logloss,2)}")
```

## ROC Curve

```python
# take the class names
    # for each class name compute y_class_actual belongs to {1,0}
    # for each class name compute y_class_proba 
    # compute fpr, tpr, threshoolds from roc_curve
    # get roc_auc from auc()
    # display the curve
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,10)
```

```python
class_names=y_actual.unique()
colors=['red','blue','green','yellow']
for i,c in enumerate(class_names):
    y_class_label=[1 if i==c else 0 for i in y_actual]
    y_class_proba=results_df[f'{c}_prob']
    fpr, tpr, thresholds =roc_curve(y_true=y_class_label, y_score=y_class_proba)
    roc_auc = auc(fpr, tpr)    
    plt.title('Receiver Operating Characteristic')    
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc,color=colors[i])
    plt.legend(loc = 'lower right')
    #plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
plt.show()
```
